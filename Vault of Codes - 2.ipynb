### Task 1:
Objective : Write a Python Code to Count word Frequency in a Given Text

```python
def count_word_frequency(text):  #Defination of a function
    words=text.split()  #Split the Text into Words
    word_freq={}  #Empty Dictionary to store Word Frequencies
    
    for i in words:  #Iteration through the list of Words
        i=i.lower()  #Convert the Word to Lowercase
        
        if i.isalpha():  #Checks whether Word consists of only Alphabetic Characters
            
            #Update the Word Frequency Dictionary
            if i in word_freq:
                word_freq[i]+=1
            else:
                word_freq[i]=1
                
    return word_freq  #Returning the Final Word Frequency Dictionary

#Driver Code
text=input("Enter your Text : ")  #Input of the Text
frequency=count_word_frequency(text)  #Calling the required Function

#Printing the Result
for i,freq in frequency.items():
    print(f"{i} : {freq}")
```

### Task 2:
Objective: Mini project

Develop a Basic to-do list Program using Functions and Data Structures.

Add Features like Adding Tasks in the to-do list, display the tasks and quitting the loop.

```python
#Initialize an Empty List to store Tasks
tasks=[]

#Function to Add  a Task to the List
def add_task(task):
    tasks.append(task)  #Inserts the Task to the List
    print(f"Task {task} Added!!!")
    
#Function to Remove a Task from the List
def remove_task(index):
    if 1<=index<=len(tasks):
        removed_task=tasks.pop(index-1)  #Removes the Specified Task from the List
        print(f"Task {removed_task} Removed!!!")
    else:
        print("Invalid Index. Please Try Again!!!")

#Function to Display the Tasks in the Lists       
def display_task():
    if not tasks:
        print("No Tasks in the List!!!")
    else:
        print("\nTasks\n")
        for index,task in enumerate(tasks,start=1):
            print(f"{index} : {task}")
        
#Driver Code
while True:
    print("\nMenu")
    print("1. Add a Task \n2. Remove a Task \n3. Display the Tasks \n4. Quit")
    
    choice=input("Enter your Choice (1/2/3/4) : ")
    if choice=='1':
        task=input("Enter your Task : ")
        add_task(task)
    elif choice=='2':
        index=int(input("Enter the Index of the Task to be Removed : "))
        remove_task(index)
    elif choice=='3':
        display_task()
    elif choice=='4':
        print("GoodBye!!!, The Loop has Quitted")
        break
    else:
        print("Invalid Choice. Please Try Again")
```
